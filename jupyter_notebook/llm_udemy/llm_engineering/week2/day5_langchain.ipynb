## Tools

Tools are an incredibly powerful feature provided by the frontier LLMs.

With tools, you can write a function, and have the LLM call that function as part of its response.

Sounds almost spooky.. we're giving it the power to run code on our machine?

Well, kinda.

```python
# Let's start by making a useful function

ticket_prices = {"london": "$799", "paris": "$899", "tokyo": "$1400", "berlin": "$499"}

def get_ticket_price(destination_city):
    print(f"Tool get_ticket_price called for {destination_city}")
    city = destination_city.lower()
    return ticket_prices.get(city, "Unknown")
```

```python
get_ticket_price("London")
```

```python
# There's a particular dictionary structure that's required to describe our function:

price_function = {
    "name": "get_ticket_price",
    "description": "Get the price of a return ticket to the destination city. Call this whenever you need to know the ticket price, for example when a customer asks 'How much is a ticket to this city'",
    "parameters": {
        "type": "object",
        "properties": {
            "destination_city": {
                "type": "string",
                "description": "The city that the customer wants to travel to",
            },
        },
        "required": ["destination_city"],
        "additionalProperties": False
    }
}
```

```python
# And this is included in a list of tools:

tools = [{"type": "function", "function": price_function}]
```

## Getting OpenAI to use our Tool

There's some fiddly stuff to allow OpenAI "to call our tool"

What we actually do is give the LLM the opportunity to inform us that it wants us to run the tool.

Here's how the new chat function looks:

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)

    if response.choices[0].finish_reason=="tool_calls":
        message = response.choices[0].message
        response, city = handle_tool_call(message)
        messages.append(message)
        messages.append(response)
        response = openai.chat.completions.create(model=MODEL, messages=messages)
    
    return response.choices[0].message.content
```

```python
# We have to write that function handle_tool_call:

def handle_tool_call(message):
    tool_call = message.tool_calls[0]
    arguments = json.loads(tool_call.function.arguments)
    city = arguments.get('destination_city')
    price = get_ticket_price(city)
    response = {
        "role": "tool",
        "content": json.dumps({"destination_city": city,"price": price}),
        "tool_call_id": tool_call.id
    }
    return response, city
```

```python

import os
from typing import Dict, Any
from dotenv import load_dotenv
import gradio as gr
from langchain_openai import ChatOpenAI
from langchain.tools import StructuredTool
from langchain.prompts import MessagesPlaceholder
from langchain.agents import AgentExecutor
from langchain.agents.format_scratchpad import format_to_openai_function_messages
from langchain.agents.output_parsers import OpenAIFunctionsAgentOutputParser
from langchain.prompts import ChatPromptTemplate
from langchain.schema.messages import HumanMessage, SystemMessage, AIMessage
from langchain_core.utils.function_calling import convert_to_openai_function

# Initialize environment
load_dotenv()

# Model configuration
MODEL = "gpt-4o-mini" # Adjust based on your needs

# Ticket prices database
ticket_prices = {
    "london": "$799",
    "paris": "$899", 
    "tokyo": "$1400",
    "berlin": "$499"
}

# Define the tool function
def get_ticket_price(destination_city: str) -> str:
    """Get the price of a return ticket to the destination city."""
    print(f"Tool get_ticket_price called for {destination_city}")
    city = destination_city.lower()
    return ticket_prices.get(city, "Unknown")

# Create LangChain tool
ticket_price_tool = StructuredTool.from_function(
    func=get_ticket_price,
    name="get_ticket_price",
    description="Get the price of a return ticket to the destination city. Call this whenever you need to know the ticket price, for example when a customer asks 'How much is a ticket to this city'"
)

# Create the LLM
llm = ChatOpenAI(
    temperature=0,
    model=MODEL
)

# Create the prompt template
SYSTEM_MESSAGE = """You are a helpful assistant for an Airline called FlightAI. 
Give short, courteous answers, no more than 1 sentence. 
Always be accurate. If you don't know the answer, say so."""

prompt = ChatPromptTemplate.from_messages([
    SystemMessage(content=SYSTEM_MESSAGE),
    MessagesPlaceholder(variable_name="chat_history"),
    HumanMessage(content="{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

# Create the agent
tools = [ticket_price_tool]
llm_with_tools = llm.bind(
    functions=[convert_to_openai_function(t) for t in tools]
)

agent = (
    {
        "input": lambda x: x["input"],
        "chat_history": lambda x: x["chat_history"],
        "agent_scratchpad": lambda x: format_to_openai_function_messages(
            x["intermediate_steps"]
        ),
    }
    | prompt
    | llm_with_tools
    | OpenAIFunctionsAgentOutputParser()
)

# Create the agent executor
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Chat function for Gradio
def chat(message: str, history: list) -> str:
    # Convert history to the format LangChain expects
    chat_history = []
    for msg in history:
        if isinstance(msg, dict):  # New message format
            if msg["role"] == "user":
                chat_history.append(HumanMessage(content=msg["content"]))
            elif msg["role"] == "assistant":
                chat_history.append(AIMessage(content=msg["content"]))
        else:  # Legacy tuple format
            human, ai = msg
            chat_history.extend([
                HumanMessage(content=human),
                AIMessage(content=ai)
            ])
    
    # Run the agent
    result = agent_executor.invoke({
        "input": message,
        "chat_history": chat_history
    })
    
    return result["output"]

# Create and launch the Gradio interface
interface = gr.ChatInterface(
    fn=chat,
    type="messages",  # Use the new message format
    title="FlightAI Assistant",
    description="Ask about flight prices and other airline information"
)


# if __name__ == "__main__":
#     interface.launch(server_name="0.0.0.0", server_port=7860)
    
```

```python
import os
import json
from fastapi import APIRouter, Response
from typing import List, TypedDict, Optional, Dict, Literal
from langchain_openai import ChatOpenAI
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain.tools import Tool
from langchain.chains import LLMMathChain
from langchain.prompts import PromptTemplate, ChatPromptTemplate


MODEL = "gpt-4o-mini"
llm = ChatOpenAI(model="gpt-4", temperature=0.0)

ticketing_tool = Tool(
    name="get_ticket_price",
    func=get_ticket_price,
    description="Get the price of a return ticket to the destination city. Call this whenever you need to know the ticket price, for example when a customer asks 'How much is a ticket to this city'",
)
tools = [ticketing_tool]
llm_with_tools = llm.bind_tools(tools=tools)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a travel agent."),
        (
            "human",
            """Previous history:
            {history}

            Question: {question}
            Please provide the flight rates back""",
        ),
    ]
)

question = "How much is a ticket to London?"
history = []
chain = prompt | llm_with_tools

# First, get the initial response
response = chain.invoke({"question": question, "history": history})

# Execute the tool and get the result
tool_result = None
if response.tool_calls:
    tool_call = response.tool_calls[0]
    tool_result = get_ticket_price(tool_call['args']['__arg1'])

# Get the final response with the tool result
final_response = llm.invoke(f"The ticket price to London is {tool_result}. Please respond to the customer's question about the ticket price.")

# Calculate API costs (in USD)
input_cost = (response.usage_metadata['input_tokens'] / 1000) * 0.03
output_cost = (response.usage_metadata['output_tokens'] / 1000) * 0.06
total_cost = input_cost + output_cost

# Pretty print all the details
print("\n=== Initial Question ===")
print(f"Question: {question}")

print("\n=== Tool Execution ===")
print(f"Tool Used: get_ticket_price")
print(f"Tool Input: London")
print(f"Tool Result: {tool_result}")

print("\n=== Final Response ===")
print(f"Response: {final_response.content}")

print("\n=== Usage and Cost Analysis ===")
print(f"Input Tokens: {response.usage_metadata['input_tokens']}")
print(f"Output Tokens: {response.usage_metadata['output_tokens']}")
print(f"Total Tokens: {response.usage_metadata['total_tokens']}")
print(f"\nCost Breakdown:")
print(f"Input Cost: ${input_cost:.4f}")
print(f"Output Cost: ${output_cost:.4f}")
print(f"Total Cost: ${total_cost:.4f}")
```

```python
import os
import json
from typing import List, Optional
from langchain_openai import ChatOpenAI
from langchain.tools import Tool
from langchain.prompts import ChatPromptTemplate
from IPython.display import display, clear_output
import time

# Initialize LLM with streaming
llm = ChatOpenAI(
    model="gpt-4",
    temperature=0.0,
    streaming=True  # Enable streaming
)

# Non-streaming LLM for initial response
llm_no_stream = ChatOpenAI(
    model="gpt-4",
    temperature=0.0,
    streaming=False
)

ticketing_tool = Tool(
    name="get_ticket_price",
    func=get_ticket_price,
    description="Get the price of a return ticket to the destination city. Call this whenever you need to know the ticket price, for example when a customer asks 'How much is a ticket to this city'",
)
tools = [ticketing_tool]
llm_with_tools = llm_no_stream.bind_tools(tools=tools)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a travel agent."),
    (
        "human",
        """Previous history:
        {history}

        Question: {question}
        Please provide the flight rates back""",
    ),
])

chain = prompt | llm_with_tools

async def process_response(question: str, history: Optional[List[str]] = None):
    """Process the response with streaming output in the notebook"""
    if history is None:
        history = []
    
    print("Getting initial response...\n")
    
    # Get initial response with tool calls
    initial_response = await chain.ainvoke({"question": question, "history": history})
    
    # Execute tool if needed
    tool_result = None
    if hasattr(initial_response, 'tool_calls') and initial_response.tool_calls:
        tool_call = initial_response.tool_calls[0]
        tool_result = get_ticket_price(tool_call['args']['__arg1'])
    
    print("Generating final response...\n")
    
    # Prepare final prompt
    destination = question.split()[-1].strip('?.')
    prompt_with_price = f"The ticket price to {destination} is {tool_result}. Please respond to the customer's question about the ticket price and suggestions for a fun family activity in the destination city."

    
    # Stream the final response
    collected_message = []
    async for chunk in llm.astream(prompt_with_price):
        if chunk.content:
            collected_message.append(chunk.content)
            # Clear previous output and display current state
            clear_output(wait=True)
            print(''.join(collected_message), end='', flush=True)
            time.sleep(0.05)  # Small delay for visibility
    

    
    return {
        'response': ''.join(collected_message),
        'tool_result': tool_result,
        'usage': {
            'total_cost': total_cost if 'total_cost' in locals() else None
        }
    }

# Example usage in notebook
# Run this in a cell:

import asyncio
question = "How much is a ticket to London?"
await process_response(question)
```

```python
# Gradio LangChain Interface
import os
import json
from typing import List, Optional
from langchain_openai import ChatOpenAI
from langchain.tools import Tool
from langchain.prompts import ChatPromptTemplate
import gradio as gr
import asyncio

# Initialize LLM with streaming
llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0.0,
    streaming=True
)

# Non-streaming LLM for initial response
llm_no_stream = ChatOpenAI(
    model="gpt-4o",
    temperature=0.0,
    streaming=False
)

# Ticket prices database
ticket_prices = {
    "london": "$799",
    "paris": "$899", 
    "tokyo": "$1400",
    "berlin": "$499"
}

# Define the tool function
def get_ticket_price(destination_city: str) -> str:
    """Get the price of a return ticket to the destination city."""
    print(f"Tool get_ticket_price called for {destination_city}")
    city = destination_city.lower()
    return ticket_prices.get(city, "Unknown")

ticketing_tool = Tool(
    name="get_ticket_price",
    func=get_ticket_price,
    description="Get the price of a return ticket to the destination city. Call this whenever you need to know the ticket price, for example when a customer asks 'How much is a ticket to this city'",
)
tools = [ticketing_tool]
llm_with_tools = llm.bind_tools(tools=tools)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a travel agent."),
    (
        "human",
        """Previous history:
        {history}

        Question: {question}
        Please provide the flight rates back""",
    ),
])

chain = prompt | llm_with_tools

async def process_response(question: str, history: Optional[List[str]] = None):
    """Process the response with streaming output"""
    if history is None:
        history = []
    
    # Get initial response with tool calls
    initial_response = await chain.ainvoke({"question": question, "history": history})
    
    # Execute tool if needed
    tool_result = None
    if hasattr(initial_response, 'tool_calls') and initial_response.tool_calls:
        tool_call = initial_response.tool_calls[0]
        tool_result = get_ticket_price(tool_call['args']['__arg1'])
    
    # Prepare final prompt
    destination = question.split()[-1].strip('?.')
    prompt_with_price = f"The ticket price to {destination} is {tool_result}. Please respond to the customer's question about the ticket price and suggestions for a fun family activity in the destination city."

    # Stream the final response
    collected_message = []
    async for chunk in llm.astream(prompt_with_price):
        if chunk.content:
            collected_message.append(chunk.content)
    
    return ''.join(collected_message)

def chat(message: str, history: List[List[str]]):
    """Gradio chat interface function"""
    # Convert Gradio history format to simple list of messages
    flat_history = [msg for pair in history for msg in pair]
    
    # Process the response asynchronously
    response = asyncio.run(process_response(message, flat_history))
    return response

# Launch Gradio interface
demo = gr.ChatInterface(
    fn=chat,
    title="Travel Agent Assistant",
    description="Ask about ticket prices and get travel suggestions!",
    examples=["How much is a ticket to London?", "What's the price for Paris?"],
    theme="soft"
)

demo.launch(server_name="0.0.0.0", server_port=7860)
```

```python
# Some imports for handling images

import base64
from io import BytesIO
from PIL import Image
```

```python
def artist(city):
    image_response = openai.images.generate(
            model="dall-e-3",
            prompt=f"An image representing a vacation in {city}, showing tourist spots and everything unique about {city}, in a vibrant pop-art style",
            size="1024x1024",
            n=1,
            response_format="b64_json",
        )
    image_base64 = image_response.data[0].b64_json
    image_data = base64.b64decode(image_base64)
    return Image.open(BytesIO(image_data))
```

# Our Agent Framework

The term 'Agentic AI' and Agentization is an umbrella term that refers to a number of techniques, such as:

1. Breaking a complex problem into smaller steps, with multiple LLMs carrying out specialized tasks
2. The ability for LLMs to use Tools to give them additional capabilities
3. The 'Agent Environment' which allows Agents to collaborate
4. An LLM can act as the Planner, dividing bigger tasks into smaller ones for the specialists
5. The concept of an Agent having autonomy / agency, beyond just responding to a prompt - such as Memory

We're showing 1 and 2 here, and to a lesser extent 3 and 5. In week 8 we will do the lot!

```python
def chat(history):
    messages = [{"role": "system", "content": system_message}] + history
    response = openai.chat.completions.create(model=MODEL, messages=messages, tools=tools)
    image = None
    
    if response.choices[0].finish_reason=="tool_calls":
        message = response.choices[0].message
        response, city = handle_tool_call(message)
        messages.append(message)
        messages.append(response)
        image = artist(city)
        response = openai.chat.completions.create(model=MODEL, messages=messages)
        
    reply = response.choices[0].message.content
    history += [{"role":"assistant", "content":reply}]

    # Comment out or delete the next line if you'd rather skip Audio for now..
    talker(reply)
    
    return history, image
```

```python
# More involved Gradio code as we're not using the preset Chat interface!
# Passing in inbrowser=True in the last line will cause a Gradio window to pop up immediately.

with gr.Blocks() as ui:
    with gr.Row():
        chatbot = gr.Chatbot(height=500, type="messages")
        image_output = gr.Image(height=500)
    with gr.Row():
        entry = gr.Textbox(label="Chat with our AI Assistant:")
    with gr.Row():
        clear = gr.Button("Clear")

    def do_entry(message, history):
        history += [{"role":"user", "content":message}]
        return "", history

    entry.submit(do_entry, inputs=[entry, chatbot], outputs=[entry, chatbot]).then(
        chat, inputs=chatbot, outputs=[chatbot, image_output]
    )
    clear.click(lambda: None, inputs=None, outputs=chatbot, queue=False)

ui.launch(inbrowser=True)
```
